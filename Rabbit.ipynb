```python
def main():
    print("Welcome to the Rabbit Name Storing Program!")

    rabbit_names = []

    while True:
        name = input("Enter the name of a rabbit (or type 'exit' to quit): ")
        if name.lower() == 'exit':
            break
        rabbit_names.append(name)
        print(f"{name} has been added to the list.")

    if rabbit_names:
        print("\nList of rabbit names:")
        for name in rabbit_names:
            print(name)
    else:
        print("No rabbit names have been entered.")

if __name__ == "__main__":
    main()
```
